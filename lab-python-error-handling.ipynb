````python
# PASO 1: initialize_inventory() con Dictionary Comprehension
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the available quantity for {product}: ")) for product in products}
    return inventory
````

````python
# PASO 1: initialize_inventory() adaptado para que funcione con validación
def initialize_inventory(products):
    inventory = {}

    for product in products:
        valid_quantity = False

        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the available quantity for {product}: "))

                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")

                inventory[product] = quantity
                valid_quantity = True

            except ValueError as error:
                print(f"Error: {error}")

    return inventory


# PASO 2: get_customer_orders() con validación de número de pedidos y producto
def get_customer_orders(inventory):
    customer_orders = set()

    valid_orders_number = False
    while not valid_orders_number:
        try:
            num_orders = int(input("Enter the number of customer orders: "))

            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative.")

            valid_orders_number = True

        except ValueError as error:
            print(f"Error: {error}")

    for i in range(num_orders):
        valid_product = False

        while not valid_product:
            try:
                product = input(f"Enter product name {i + 1}: ").strip()

                if product not in inventory:
                    raise ValueError("This product is not in the inventory.")

                if inventory[product] <= 0:
                    raise ValueError("This product is out of stock.")

                customer_orders.add(product)
                valid_product = True

            except ValueError as error:
                print(f"Error: {error}")

    return customer_orders


# PASO 3: calculate_total_price() con validación del precio
def calculate_total_price(customer_orders):
    total_price = 0

    for product in customer_orders:
        valid_price = False

        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))

                if price < 0:
                    raise ValueError("Price cannot be negative.")

                total_price += price
                valid_price = True

            except ValueError as error:
                print(f"Error: {error}")

    return total_price


# PASO 4: update_inventory() para restar stock y eliminar productos con 0
def update_inventory(customer_orders, inventory):
    updated_inventory = {}

    for product, quantity in inventory.items():
        if product in customer_orders:
            quantity -= 1

        if quantity > 0:
            updated_inventory[product] = quantity

    return updated_inventory


# PASO 5: calculate_order_statistics() para calcular estadísticas del pedido
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    order_statistics = (total_products_ordered, percentage_ordered)
    return order_statistics


# PASO 6: print_order_statistics() para imprimir estadísticas
def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print(f"Total products ordered: {order_statistics[0]}")
    print(f"Percentage of unique products ordered: {order_statistics[1]}%")


# PASO 7: print_updated_inventory() para imprimir inventario actualizado
def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")


# PASO 8: ejecución principal del programa
products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)
total_price = calculate_total_price(customer_orders)
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)

print(f"Total price of the customer order: {total_price}")
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
````
